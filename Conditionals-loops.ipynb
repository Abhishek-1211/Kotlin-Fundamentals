# 'if' Expression

can be used to return a value

```kotlin
var a = 5
var b = 10
if(b>a)
println("b is greater")
else
println("a is greater")

var ans = if(b>a)
println("b is greater")
else
println("a is greater")

print(ans)
```

```kotlin
var check = if(10>5)"yes" else "no"
println(check)
print(check.javaClass)
```

```kotlin
var checkAgain = if(10<5)"String" else 5
println(checkAgain)
print(checkAgain.javaClass)
```

# When Expression

```kotlin
var alarm = 16
when(alarm) {
    12 -> println("less than 15")
    in 12..18 -> println("between 12 to 18")
    else -> println("less than 15")
} 

var output = when(alarm) {
                12 -> "less than 15"
                in 12..18 -> "between 12 to 18"
                else -> "greater than 18"
              } 
print("output:"+output)
```

# For loop

```kotlin
for(i in 1 until 10) {
    print("$i ")
}

println()
for(i in 1 until 10 step 2) {
    print("$i ")
}

println()
for(i in 1..10 step 1) {
    print("$i ")
}
```

```kotlin
for(i in 10 downTo 2) {
    print("$i ")
}
println()
for(i in 10 downTo 2 step 2 ) {
    print("$i ")
}
```
