1.**Install miniconda**

check /opt folder for installed directory
(environment location: /opt/miniconda3)
    
2.**Install jupyter(on terminal)**

_conda install jupyter_
    
3.**Install kotlin kernel(on terminal)**

_conda install -c jetbrains kotlin-jupyter-kernel_
    
4.**Start jupyter notebook (from terminal)**

_jupyter notebook_
    
5.**Open New notebook**

File -> New Notebook -> Select kotlin kernel

# Variables

    Read only variables - val
    Mutable variables - var

* ####  **statically typed language** 

    check the type of variable at compile time and not runtime

* #### **supports type-inference**

    we needn't provide data type while declaring the varibles

```kotlin
var name: String = "Alex"
print("Hello $name")
```

```kotlin
name = "John"
print("Hello $name")
```

```kotlin
val username : String = "Peter"
username = "Rahul" //cannot be reassigned
print(username)
```

```kotlin
val age: Int = 22
print(age)
```
