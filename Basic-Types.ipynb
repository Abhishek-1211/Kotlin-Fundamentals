# Floating Numbers



1.default type inference is double

2.Need to explicityly specify for float with f or F

check data type of any variable

    Using ::class Property
    Using javaClass Property:

```kotlin
//default type inference is Double for floating number
var floatNumber = 2.5
println(floatNumber.javaClass)
println(floatNumber::class)

var myFloatNumber = 2.5f
print("myFloatNumber Data type is :"+myFloatNumber.javaClass)
```
