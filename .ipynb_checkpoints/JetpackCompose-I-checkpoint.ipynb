## Lazy lists
* #### to display a large number of items (or a list of an unknown length)
* #### only composes and lays out the currently visible items.
* #### Common Lazy Layouts
  * #### LazyColumn: A vertical list.
   * #### LazyRow: A horizontal list.
  * #### LazyVerticalGrid: A grid-like structure



------------


```kotlin
var data = mutableListOf<String>()

LazyColumn {
    
    // for displaying single item we use 'item' block    
    item {
        Text("Header")
    }
    
    //for displaying list of items we use 'items' block
    items(data) {
        item -> Text("$item")
    }

    //used in cases when you need to know indexes of items
    itemsIndexed(data) {
        index, item -> Text("$index:$item")
    }
}
```

#### Now we can customise list appearnce(lazy Column) using below paramters

* #### Summary of Parameters
| Parameter |	Purpose |
| :--------- |:------------|
|modifier|	Customizes appearance and layout (e.g., size, padding, background).|
|verticalArrangement|	Controls vertical spacing/alignment of items.|
|horizontalAlignment|	Aligns items horizontally within the column.|
|contentPadding|	Adds padding around the content.|
|reverseLayout|	Reverses the item order.|
|userScrollEnabled|	Enables or disables user scrolling.|
|state|	Observes/manages the scroll state.|
