# Inheritance

```kotlin
open class Vehicle (var name:String, var color:String) {

    open fun move() {
        println("$name is moving")
    }

    fun stop() {
        println("$name has stopped")
    }

}

class Car(name:String, color:String, var model:String): Vehicle(name, color) {

    override fun move(){
        println("$name car having $color color is moving")
        super.move()
    }
}

var myCar = Car("Toyota", "Grey","Corola")
myCar.move()
```

# Sealed class

* ####  used to represent a  fixed hierarchy of classes with different behaviors and properties for each subclass.In contrast enum classes are for representing a fixed set of constants or choices with simpler, uniform behavior
* #### these cannot be instantiated

```kotlin
sealed class Result(val message:String) {
    class Success(message: String):Result(message)
    sealed class Error(message:String):Result(message) {
        class RuntimeError(exception:Exception, message:String):Error(message)
        class CompiletimeError(exception:Exception, message:String):Error(message)
    }
}

fun getData(result:Result){
    when(result) {
        is Result.Success -> println("Success")
        is Result.Error.RuntimeError -> println("run-time-error")
        is Result.Error.CompiletimeError -> println("compile-time error")
    }

}
var res = Result.Success("Success")
val exception = IllegalArgumentException()
val res2 = Result.Error.RuntimeError(exception,"Error")
getData(res)
getData(res2)
```

# Abstract class

* #### sharing common behaviour across subclasses
  - #### meant to be inherited
  
  - #### cannot be instantiated

* #### can contain abstract as well non-abstract methods
* #### these classes just give overview of what to do and leave it to subclasses on how to do
* #### `unlike interfaces` they can have properties, primary constructor or secondary constructors which help to intitialise state

```kotlin
abstract class Vehicle {

    abstract fun move()
    abstract fun stop()
}

class Car(var name:String, var color:String, var model:String): Vehicle(){
    override fun move(){
        println("move():inside car")
    }

    override fun stop(){
        println("stop():inside car")
    }
}

var myCar = Car("Toyota","balack","Corola")
myCar.stop()
```

* #### application1:Data Access Layer in Database Operations

```kotlin
//Data Access Layer in Database Operations

abstract class Database {
    abstract fun connect():String
    abstract fun executeQuery(query:String):String
}

class MySQLDatabase:Database(){
    override fun connect():String{
        return "connected to MySQL DB....."
    }

    override fun executeQuery(query:String):String{
        return "query executed successfully....."
    }
}


class MongoDB:Database() {
    override fun connect():String {
        return "connected to mongo DB"
    }

    override fun executeQuery(query:String):String{
        return "query executed successfully for mongoDB....."
    }
}
    
var mysql = MySQLDatabase()
println(mysql.connect())
println(mysql.executeQuery("Select * from myDB"))

var mongoDb = MongoDB()
println(mongoDb.connect())
println(mongoDb.executeQuery("execute"))
```

# Data class


* #### these are classes designed to hold-data(store and manage information) similar to POJO classes in JAVA.
* #### But unlike Java, they automatically generates a custom implementation of  equals(),hashcode(), toString(),copy() methods based on the properties declared in the primary constructor thereby `avoiding lots of boilerplate code`

## breakdown

* #### `toString()`: Returns a string like User(firstName=Rohit, lastName = Sharma, age=25) when printing an instance.
* #### `equals()`: Compares objects based on their property values, not references, so person1 == person2 returns true even though they are different instances.
* #### `hashCode()`: Generates a hash code for the object based on its properties.
* #### `copy()`: Allows you to create a new object by copying an existing one, optionally changing some properties. Java POJOs typically don’t have this functionality out of the box, so you'd need to write your own copy method
* #### `Destructuring`: You can break down the object into its properties using destructuring declarations.


## Rules for Data Classes:

* #### `Primary constructor`: A data class must have at least one property  in the primary constructor

* #### `Cannot be abstract, open, sealed, or inner`: Data classes cannot be abstract, open (inheritable), sealed, or inner classes.

* #### `Automatic equals(), hashCode(), and toString()`: These methods are auto-generated based on the properties declared in the primary constructor. If you need custom behavior, you can override them.



<br>

<br>



```kotlin
//structural equality vs referential equality
var name1 = "Arpit"
var name2 = "Rohit"
var name3 = name1
println(name1 == name2)
println(name1 === name2)
println(name1 === name3)
```

```kotlin
class User(var firstName:String, var lastName:String, var Age:Int) {

    /*
    here we only override eqauls method of Any class- However, general
    convention is that whenever we override equals method we must also
    override hascode method.thus 2 objects will be equal only when they are equal 
    as per custom implementation of equals method alongwith same hashcode.This is 
    particularly useful in case of Hashset and HashMap  
    */
    
    override fun equals(other:Any?):Boolean {
        
        if(this===other) return true
        
        if(other is User){
            
            return (this.firstName==other.firstName 
                    && this.lastName == other.lastName
                    && this.Age == other.Age)
        }
        
        return false
    }
}

var user1 = User("Rahul", "Gupta", 28)

println(user1.firstName)

var user2 = User("Rahul", "Gupta", 28)

var user3 = User("Rohit","Sharma",34)

var user4 = User("Rohit","Sharma",34)

println(user1 == user2)
println(user1 === user2)
println(user1 == user3)
println(user4 == user3)
```

```kotlin
/*
So, with data classes we needn't override the equals method ,
hashcode, toString() etc as these are automatically genrated 
*/

data class User(var firstName:String, var lastName:String, var age:Int)


var user1 = User("Rahul", "Gupta", 28)

println(user1.firstName)

var user2 = User("Rahul", "Gupta", 28)

var user3 = User("Rohit","Sharma",34)

var user4 = User("Rohit","Sharma",34)

println(user1 == user2)
println(user1 === user2)
println(user1 == user3)
println(user4 == user3)


println(user4)
println(user4.hashCode())

println(user3.hashCode())

val user5 = user4.copy(age = 24)
println("User5:$user5")
//println("user1:hascode=$ {user1.hashcode()")

```

## Destructing

* #### allows to break down objects into indidvidual properties in concise manner.<br>Typically, you can access properties using below syntax:
  #### `user1.firstName or user1.age`
* #### Destructuring declarations allow you to extract these properties directly into variables in a single line

* #### This is possible because Kotlin automatically generates `component functions-componentN()` for data classes.

```kotlin
data class Student(var name:String, var rollNo:Int)

var student1 = Student("Prakash",25)
val(studentName,rollNumber) = student1
println(studentName)
println(rollNumber)

//component functions
println(student1.component1())
println(student1.component2())
```

## Component functions 
* #### these are functions that allow destructuring of the object’s properties. 
* #### Each componentN() function corresponds to one of the properties declared in the primary constructor of the data class, where N is the order of the property.
* #### Kotlin automatically generates the following component functions for the above student class:
  * #### component1() for the first property, name
  * #### component2() for the second property, rollNo

* #### for `Non-data classes` we have to manually define component functions to facilitate destructing



```kotlin
//typical component function
fun component1(): String {
    return name
}

fun component2(): Int {
    return rollNo
}
```

## Interfaces

* #### share common behaviour across group of classes.
* #### we define what needs to be done and leaving it to class to implement the how is to done
* #### can't be instantiated (so `can't have constructors`)

------

### Interfaces Vs Abstract class

|                 Abstarct class              |          Interface                   |
|:-------------------------------------------:                  |:-----------------------------------: |
|classes have a clear inheritance relationship and share a common identity|  classes don’t share a common identity but need to support a particular behavior             |
|  doesn't support multiple inheritance | Interfaces allow for multiple inheritance. |
|Both can't be instantiated but abstarct class can have constructors (primary as well as secondary| can't be instantiated and can't have  constructor|
|can initialize state(i.e., set up properties with values) or take parameters.|designed purely to define behavior, not to hold state or initialize objects.

----------

```kotlin
interface IEngine {
    //In interface we just define the functions
    fun startEngine()
}

class Car(var name:String, var color:String):IEngine {
    override fun startEngine(){
        println("Starting the car......")
    }
}

var car = Car("Toyota","black")
car.startEngine()

class Truck(var name:String, var color:String, var model:String):IEngine{
    override fun startEngine(){
        println("Starting the Truck......")
    }
}

var myTruck = Truck("Swaraj","Red","I6578")
myTruck.startEngine()
```

## Object expressions

* #### way to create an anonymous object or an instance of a class without explicitly defining a new class
* #### typically used when you need a local, short-lived object for a particular purpose.
* #### can be used to implement interfaces/extend classes

### Use cases
* #### __Quickly Define Behavior__: Used when you need to quickly define a new behavior or functionality that’s specific to a local context, without creating a new class.
* #### __Event Handling__: Commonly used for event listeners or callbacks, where you need to pass an object that implements an interface.
* #### __Local Customization__: When you need a one-off customization of a class for a specific purpose

```kotlin
var myObject = object{
    var x = 15
    fun printX(){
        println(x)
    }
}
```

```kotlin
interface onClickListener {
    fun onClick()
}

class clickListener:onClickListener{
    override fun onClick(){
    }
}
class Button(var text:String, var id:Int, onClickListener:onClickListener)

var loginButton = Button("login",12341,object: onClickListener{
    override fun onClick() {
        println("login button clicked")
    }
})
```
