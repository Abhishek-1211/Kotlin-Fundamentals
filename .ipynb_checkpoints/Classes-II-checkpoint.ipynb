# Classes and Objects

## Enum classes
 * #### special type of class used to represent a group of constants, known as enumerations. 
 * #### Each constant in an enum class is an object and is considered an instance of the enum class

### Key Points
* #### __Fixed set of constants__: Enum classes are ideal when you need to represent a set of predefined constants or values.
* #### __Each constant is an object__: Each enum constant is a `singleton object`, meaning it represents one instance of the enum class.
* #### __Can have properties and methods__: Enum classes can also include properties, methods, and even implement interfaces, making them more powerful than basic enums found in some other programming languages.
* #### __Type-safe__: Enums in Kotlin are type-safe, so you cannot assign an invalid value to a variable of an enum type.


```kotlin
enum class Direction {
    NORTH,
    SOUTH,
    EAST,
    WEST,
}

println(Direction.NORTH)
for (direction in Direction.values()){
    println("${direction.ordinal} $direction" )
}
```

## Inner classes

* #### classes which are nested inside another class
* #### used when there is close relationship between 2 classes

```kotlin
class ListView(val items:Array<String>) {
    inner class ListViewItem {
        fun displayItem(position:Int){
            println(items[position])
        }
    }
}

var listView = ListView(arrayOf("item1","item2","item3","item4"))
listView.ListViewItem().displayItem(2)
```

## challenge
* #### create class to represent bank account of a person.
* #### class to have 3 properties - accountname,balance and transactions(mutable list)
* #### class to have 2 functions for deposit , withdraw and calculating the balance alongwith with validation checks like you cannot withdraw when balance<0 etc.

```kotlin
class Account(var accountName:String) {

    var balance = 0 
    var transactions = mutableListOf<Int>()
    
    fun deposit(amount:Int) {
        transactions.add(amount)
        balance += amount
    }
    
    fun withdraw(amount:Int) {
        if(balance ==0 || balance < amount){
            println("Insufficient balance")
            return
        }
        transactions.add(-amount)
        balance -= amount 
    }
    
    fun checkBalance() {
        println(this.balance)
    }
}

var account = Account("MySavingsAccount")
println(account.balance)
account.deposit(500)
account.deposit(500)
account.deposit(500)
println(account.transactions)
account.deposit(500)
account.withdraw(450)
println(account.transactions)
account.checkBalance()
account.withdraw(2050)
account.withdraw(450)
account.checkBalance()
```
