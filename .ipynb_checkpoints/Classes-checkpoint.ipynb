## `lateinit` keyword
> - #### used to decalre a non-nullable property that will be initialised later
> - #### typically used when u don't want/cannot initialize a  property at time of declaration and also avoid using nullable types(?)

### Key Points:
> 1. #### __Used with var properties__: can only use lateinit with `var`, not val, because lateinit properties need to be mutable
> 1. #### __non-nullable__: It works with non-nullable types, meaning you don't have to declare the property as nullable (no need for ?).
> 2. #### __Deferred Initialization__: The property will be initialized at a later point, not during the class instantiation.
> 3. #### __Error if Accessed Before Initialization__: If you try to access a lateinit property before it’s initialized, Kotlin will throw an exception (`UninitializedPropertyAccessException`).


### Limitations of `lateinit`:
> 1. #### __Only for `var` properties__: You cannot use lateinit with val because val properties are immutable, and lateinit implies deferred initialization.
> 1. #### __Primitive Types__: lateinit is not allowed with primitive types like Int, Float, Double, etc. You need to use nullable types or default values instead.

### Common Use cases:
> 1. #### __Dependency Injection__:
>    #### In a framework where objects are injected later (e.g., Dagger or Spring), lateinit is ideal because the object cannot be initialized at the point of class creation
> 3. #### __Android Developement__:
>    #### In Android, UI components like TextView or Button can only be initialized after the onCreate() or onViewCreated() methods, and lateinit is often used to avoid nullable types
> 5. #### __Unit Testing__:
>    #### In testing scenarios, you may not have real objects available at the time of class creation, but mock objects will be injected later. lateinit is commonly used in such cases.

```kotlin
class User(var firstName:String, var lastName:String, var age:Int) {
    var favouriteMovie
    
}
```

```kotlin
//compiles successfully
class User(var firstName:String, var lastName:String, var age:Int) {
    lateinit var favouriteMovie:String
    
}
```

## Companion Object

- #### these are objects for directly accessing properties and methods without creating an instance of class

### Key Points:
1. #### __Belongs to the class__: A companion object is tied to the class, not to the instances of the class.
2. #### __Acts like a singleton__: Each class can have `one companion object`, which means it's a single instance shared across all uses of the class
3. #### __Access via class name__: Members of the companion object can be accessed using the class name, similar to how you would access static members in Java.
4. #### __Implements interfaces__: A companion object `can implement interfaces`, which makes it more powerful than static methods in Java.


```kotlin
class Calculator() {
    companion object{
        fun sum(num1:Int, num2:Int) = num1 + num2
    }
}

print(Calculator.sum(5,6))
```

## Singleton class

### Key Points:
1. #### Single Instance: A singleton guarantees that only one instance of the class exists throughout the entire application.
2. #### Global Access: You can access the singleton directly using its name, without needing to instantiate it.
3. #### Thread-Safe: The object keyword makes the singleton automatically thread-safe in Kotlin.
4. #### Shared State: Singleton objects can hold shared state and behavior across different parts of the application.


### Use Cases for Singleton:
1. #### __Configuration Manager__: A singleton can hold global configuration settings for an application.
2. #### __Logger__: Maintaining a single logging instance to write logs to a file or console.
3. #### __Database Connection__: Managing a single shared connection to a database.
4. #### __Cache__: A shared cache instance used across the application.
5. #### __Service__: Singleton services that need to maintain shared state or behavior.

```kotlin
object Database {
    init {
        println("object created")
    }
}

println(Database)
println(Database)
```

```kotlin
object Logger {
    
    var logs:MutableList<String> = mutableListOf()
    fun log(message:String){
        logs.add(message)
        println("logs added: $message")
    }

    fun printlogs(){
        for(log in logs){
            println(log)
        }
    }
}

Logger.log("Application started...")
Logger.log("fetcing configuration settings:...")
Logger.log("checking port availability...")
Logger.log("checking disk space...")

Logger.printlogs()
```

```kotlin
object Cache {
    
    var cacheMap = mutableMapOf<String,Any>()

    // add data into cache
    fun addData(key:String,value:Any) {
        cacheMap[key]=value
        println("Added data to cache:key-$key and Value-$value")
    }
    //retrieve data from cache
    fun getData(key:String):Any? {
        return cacheMap[key]
        //println("Added data to cache:key-$key and Value-$value)
    }
    
    // handle cache hit and miss
    fun contains(key:String):Boolean {
        return cacheMap.containsKey(key)
    }
    // remove data from cache
    //clear the cache
    
}

Cache.addData("India","New Delhi")
Cache.addData("Srilanka","Colombo")
println(Cache.cacheMap)
println(Cache.getData("India"))
println(Cache.getData("Srilanka"))
println(Cache.contains("South-Africa"))
```

## lazy initialisation

* #### Lazy initialization defers the creation of an object until it is actually needed, improving performance and resource usage.
* #### The object is initialised only once, when  it is accessed for first time

### When to Use Lazy Initialization:
#### Lazy initialization is useful in scenarios like:

1. #### __Expensive Resource Allocation__: When the initialization involves heavy operations, like loading a large dataset or establishing a network connection.
2. #### __Conditional Use__: When the object may not be used in every program execution. For example, certain objects may only be needed based on user input or specific runtime conditions.
3. #### __Large Applications__: In large applications where start-up time is critical, deferring initialization can significantly improve performance.

```kotlin
class User (var firstName:String, var lastName:String) {
    init {
        println("object $firstName has been created")
    }
}

var user1 = User("Rahul", "Kanwar")
//var user2 = User("User2", "Kumar")
var user2 by lazy {
    User("user2","kumar")
}
```

```kotlin
class User (var firstName:String, var lastName:String) {
    init {
        println("object $firstName has been created")
    }
}

fun main() {
    
    var user1 = User("Rahul", "Kanwar")
    //var user2 = User("User2", "Kumar")
    val user2 by lazy {
        User("user3","kumar")
    }
}

main()
```
