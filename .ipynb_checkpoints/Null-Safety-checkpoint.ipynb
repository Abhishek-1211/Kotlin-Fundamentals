# Null Saftey

1.By default all objects and variables are non-nullable

2.IF you want a variable to be null, you need to explicity specify using '?' operator

```kotlin
var b:String? = "abc" // can be set to null
b = null // ok
print(b)
```

```kotlin
var myName:String? = "Rhitwik"
print(myName.length)
//error-Line_5.jupyter.kts (2:13 - 14) Only safe (?.) or non-null asserted (!!.) calls are allowed on a nullable receiver of type String?
```

## safe call operator

```kotlin

var newName:String? = "geeta"
//'?.'-safe call operator will check for null value -if reference variable is not null go ahead else output null
println(newName?.length)

//however,if you want the compiler to throw null pointer exception in case value goes null use -'!!' operator
//not-null assertion operator (!!) converts any value to a non-nullable type and throws an exception if the value is null
newName = null
print(newName!!.length)
```

```kotlin
var myVar:String? = "Hello"
print(myVar?.length)
myVar = null
print(myVar!!.length)
```

## Elvis operator

- Modification of safe call operator.

- If you want to output some non-null value in case reference variable goes null use '?:' operator


```kotlin
var checkElvis: Int? = 6
checkElvis = null
print(checkElvis ?: "output")
```

```kotlin
var myAge:Int? = 30
print(myAge?.minus(20)?:"sorry")

myAge = null
print(myAge?.minus(20)?:"\nsorry")
```

### Practice Questions

```kotlin
fun feedback(message:String?){
    print(message?:"No feedback provided")
}

feedback("Great service!")
```

```kotlin
fun checkProductPrice(price:Int?):String{
    return if(price!=null){
        if(price > 100) "Premium Product"
        else "Standard Product"
    }
    else
    "Price not available"
}

checkProductPrice(123)
```

```kotlin
fun checkProductPrice(price:Int?):String{
    return price
            ?.let { if(it>100) "Premium Product" else "Standard Product"}
            ?:"Price not available"
}

checkProductPrice(123)
```

```kotlin
fun checkProductPrice(price:Int?):String{
    return if(price!=null){
        if(price > 100) "Premium Product"
        else "Standard Product"
    }
    else
    "Price not available"
}

checkProductPrice(123)
```

```kotlin
fun processRating(rating:Int?):String {
    return when(rating){
        null -> "No rating provided"
        in 1..5 -> "Rating:$rating"
        else -> "invalid rating"
    }
}


processRating(7)
```

```kotlin
fun formatAddress(address:String?):String {
    //return if(address!=null)"Address:$address" else "Address not provided"

    return address?.let{"$it"}?:"Address not provided"
}

formatAddress("123,MG Road,Civil Lines, Delhi")
```

```kotlin
fun getUserInfo(user:String?):String {
    return user?.let{"User: $user"}?:"User not found"

}

getUserInfo("John")
```

```kotlin
fun validateEmail(email:String?):String {
    return when {
        email == null -> "No input provided"
        email.contains("@")-> "valid email"
        else -> "invalid email"
    }
}

validateEmail("abvc@gmail.com")
```
