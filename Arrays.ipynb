```kotlin
var mixedArray = arrayOf(1,2,3,4,5,"Hello",'c')
for(item in mixedArray){
    print("$item  " )
}
println()

println("The size of array is: ${mixedArray.size}")


for(item in mixedArray){
    if(item is Int) {
        print("$item ")
    }
}
println()
print(mixedArray::class)
```

```kotlin
var countries:Array<String> = arrayOf ("India","Srilanka","Russia","England","Pakistan")

countries.forEach {
    println(it)
}
```
