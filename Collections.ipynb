# Collections

* #### Kotlin collections are mainly divided into three types:

    * #### __List__: An ordered collection of elements that allows duplicates.
    * #### __Set__: An unordered collection of unique elements.
    * #### __Map__: A collection of key-value pairs, where each key is unique.
  
* #### Collections in Kotlin are either mutable or immutable:

    * #### Immutable collections: Cannot be modified after creation (e.g., List, Set, Map). These are ideal for read-only scenarios.
    * #### Mutable collections: Can be modified (e.g., MutableList, MutableSet, MutableMap). They allow adding, removing, or updating elements.

```kotlin
var names = mutableListOf("Rahul","Geeta","Avinash")

names.forEach {
    println(it)
}

names.add(1,"Sheetal")

println("--------New list--------------")

names.forEach {
    println(it)
}
println()
println(names)
names.forEach(::println)
```

## Transformation in Map
### `mapKeys`
* #### mapKeys transforms only the keys in a map while keeping the values unchanged.
* #### It takes a lambda that allows you to define how each key in the map should be transformed, returning a new map with the transformed keys.

### `mapValues`
* #### mapValues transforms only the values in a map, keeping the keys unchanged.
* #### It takes a lambda that allows you to define how each value should be transformed based on the current value (and the key if needed).

### Imp Points:
* ### Immutability:

    * #### The mapKeys/mapValues function does not modify the original map.Instead, it returns a `new map` with the transformed keys.If you want to update numbersMap with the new keys, you need to reassign it:

* ### Preserving Mutability:

    #### Since mapKeys returns an immutable Map, you might need to convert it back to a MutableMap using `.toMutableMap()` if you intend to modify it further.


```kotlin
var numbersMap = mutableMapOf("key 1" to 1,"key 2" to 2, "key 3" to 3, "key 4" to 4)


println(numbersMap.mapKeys{entry -> entry.key.uppercase()})

println(numbersMap.mapValues{entry -> entry.value * 10})

numbersMap = numbersMap.mapValues{entry -> entry.value * 10}.toMutableMap()
numbersMap
```

## Zipping

* #### combine two collections into a single collection of pairs,
* #### herein each element from the first collection is paired with the corresponding element in the second collection
* #### we also have `unzip` function to unzip a list of pairs

```kotlin
var names = mutableListOf("Abhishek","Sumit","Parth","Kamal")
var ages = mutableListOf(30,22,23,61)

println(names.zip(ages))
```

#### Transformation with Zip

```kotlin
var names = mutableListOf("Abhishek","Sumit","Parth","Kamal")
var ages = mutableListOf(30,22,23,61)

println(names.zip(ages){name,age->"$name 's age is $age"})
```

```kotlin
var numberPairs = listOf("one" to 1, "two" to 2, "three" to 3)
println(numberPairs)
println(numberPairs.unzip())
```

## Association
* #### used to create maps form collections
* ### __associate__
    #### __Purpose__: Creates a map where `each element is transformed` into a key-value pair using a given function.
    #### __Usage__: associate is used when `both the key and value for each map entry are derived from each element` in the collection

* ### __associateBy__
    #### __Purpose__: Creates a map where `each element itself becomes the value` in the resulting map, and the key is derived from a specific property or transformation of each element.

```kotlin
var names = listOf("Rahul","Amrita", "Beghar","Bechara")
println(names.associate{it -> it to it.length})
println(names.associateWith{it.length})

println(names.associateBy{it.first().uppercase()})
```

## flatten
* #### useful when you have lists of lists (or similar nested structures) and want to create a single list with all the elements in a flat structure.
* #### flatten can be applied to collections like List<List<T>>, Set<Set<T>>, Sequence<Sequence<T>>, or any other collection where elements are themselves collections.
* #### It returns a single list with all elements from the nested collections, discarding the original structure of the nested collections.

```kotlin
var nestedList = listOf(listOf(1,2,3),listOf(4,5,6))
var flattenedList = nestedList.flatten()
flattenedList
```

```kotlin
var nestedList = listOf(listOf(listOf(1,2,3),listOf(4,5,6)),listOf(7,8,9))
var flattenedList = nestedList.flatten()
flattenedList
```

## String Representation

| `Function`  |   `Description`  |

## Filtering

* #### to operate filtering result you should assign it to a variable or chain the function after filtering(The original collection remains unchanged after filtering. So, `If you don’t store or chain the result`, the filtered collection will be discarded, and any further operations on it won’t have an effect)


### Common Filtering functions

|Functions|Description|
  |:----------|:--------|
  |filter|filter takes a lambda predicate as a parameter and returns a new collection with elements that satisfy the condition.|
  |filterNot|filterNot returns elements that do not satisfy the predicate.|
  |filterIsInstance|filterIsInstance filters elements of a specific type from a collection|
  |filterNotNull|filterNotNull removes all null elements from a collection of nullable types.|
  |filterIndexed|filterIndexed lets you use both the index and the value in the filtering predicate. It’s useful when the position of elements is significant in the filtering logic.|
  |filterKeys and filterValues(for Maps) |For maps, filterKeys filters based on the keys, while filterValues filters based on the values.|


------------

```kotlin
var students = listOf("Sachin", "Rajesh","Atul wassan","Premjit")
var lengthyNames = students.filter{it.length>6}
lengthyNames
```

```kotlin
var keyMap = mapOf("Key 1" to 1, "key 2" to 2, "key 3" to 3, "key 4" to 4)
var filteredMap = keyMap.filter{it.value >= 3}
filteredMap
```

```kotlin
var mixedList = listOf(1,2,3,4,5,6,"Hello","how","are","you",true,false,true)
var booleans = mixedList.filterIsInstance<Boolean>()
println(booleans)

var filteredString = mixedList.filterIsInstance<String>()
println(filteredString)

var numbers = mixedList.filterIsInstance<Int>()
val(even,odd) = numbers.partition{it%2==0}
println(even)
println(odd)
```

## Grouping functions 
* #### these functions are  used to group elements of a collection based on a certain criterion.
* #### `returns a map` where each key represents a group, and the value is a list of elements in that group. 
* #### Grouping functions help organize collections by shared attributes or conditions, which can simplify data analysis or transformation tasks.



### Key Grouping Functions
1. #### __groupBy__
    #### groupBy takes a lambda (or key selector function) to determine the grouping key and returns a Map where each key maps to a list of items in that group.
   1. #### groupBy with Key Selector and Value Selector:
        #### This version takes both a key selector and a value selector. It allows you to control the values in each group by applying a transformation to each element.



```kotlin
var numbersStrings = listOf("one","two","three","four","five","six")
println(numbersStrings.groupBy{it.first().uppercase()})

//overloaded version of groupBy 
println(numbersStrings.groupBy({it.first()},{it.uppercase()}))
```

## Comparable and Comparator

* #### Comparable is used when you need a single universal sorting logic for any class.Sorting logic is implemented within the class itself
* #### The Comparator interface allows you to define custom, external ordering for classes `without modifying` the class itself. This is especially useful when:
    1. #### Multiple sorting criteria are needed for a class.
    2. #### Sorting logic needs to be applied externally (like in third-party libraries or immutable classes).

```kotlin
data class Laptop(val brand:String, val year:Int, val ram:Int, val price:Int): Comparable<Laptop> {
    
    override fun compareTo(other:Laptop):Int {
        if(this.price>other.price) return 1
        else if(this.price<other.price) return -1
        else return 0
    }
}

val laptops = mutableListOf(
    Laptop("Dell",2021,8,34000),
    Laptop("lenovo",2021,8,24000),
    Laptop("HP",2021,8,32000),
    Laptop("MAC",2021,8,65000)
    )

println(laptops.sorted().forEach{println(it)})
```

```kotlin
data class Laptop(val brand:String, val year:Int, val ram:Int, val price:Int)
val laptops = mutableListOf(
    Laptop("Dell",2016,8,34000),
    Laptop("lenovo",2017,4,24000),
    Laptop("HP",2021,16,62000),
    Laptop("MAC",2022,8,65000)
    )

class ComparatorRam: Comparator<Laptop>{
    override fun compare(laptop1:Laptop,laptop2:Laptop):Int{
         if(laptop1.ram > laptop2.ram) return 1
        else if(laptop1.ram <laptop2.ram) return -1
        else return 0
    }
}

laptops.sortedWith(ComparatorRam()).forEach{println(it)}

println("--------------------------")
//shorter way 
laptops.sortedWith(compareBy{it.year}).forEach{println(it)}
println("--------------------------")

//more concise way
laptops.sortedBy{it.price}.forEach{println(it)}
```
