````kotlin
fun getMax(a:Int,b:Int):Int {
    var max = if(a>b) a else b
    return max
}

print(getMax(5,9))
````

````kotlin
//we can use return statment without returning anything -but in this function, should also not have any return type
fun getMax(a:Int,b:Int) {
    var max = if(a>b) a else b
    return
}
````

````kotlin
fun getMax(a:Double,b:Double):Int{
    var max = if(a>b) a else b
    return max.toInt()
}
println(getMax(5.0,12.0))


//we can also rewrite above function as it conatins only one statement
fun getMax2(a:Double,b:Double) = if(a>b) a else b

print(getMax2(5.0,12.0))
````

## Funtion overloding


````kotlin
fun getMax(a:Int,b:Int):Int {
    var max = if(a>b) a else b
    return max
}

fun getMax(a:Double,b:Double):Double{
    var max = if(a>b) a else b
    return max
}

fun getMax(a:Int,b:Int,c:Int):Int {
    var max = if(a>b) if(a>c) a else c else if(b>c) b else c
    return max
}

println(getMax(8,10))
println(getMax(8.0,10.0))
println(getMax(8,10,20))
````

## Default parameters

````kotlin
fun showMessage(greeting:String ="Hello", name:String="User") {
    println("$greeting, $name")
}

showMessage()
showMessage("Welcome")
showMessage("Welcome","Rahul")
showMessage(name = "Kamal")
````

### Varargs

````kotlin
fun sumNumbers(vararg numbers:Int):Int {
    var res = 0
    for(number in numbers){
        res += number
    }

    return res
}

println(sumNumbers(1,2,3,5,18,19))
println(sumNumbers(1,2,3,5,18,19,20,24,26))
````

````kotlin
fun sumNumbers(vararg numbers:Int) {
    numbers.forEach {
        print("$it ")
    }
}

sumNumbers(1,3,5,6)
````

````kotlin
fun capital(countryName:String) : String{
    var capitalCity =""
    when(countryName) {
        "India"->capitalCity = "New Delhi"
        "Pakistan"-> capitalCity = "Islamabad"
        "Srilanka"-> capitalCity = "Colombo"
        else -> capitalCity = "No matches found"
    }
    return capitalCity
}

var capitalCity = capital("India")
println(capitalCity)
println(capital("Russia"))
````

## Scope Functions
* #### Scope functions are special functions in Kotlin (like __let__, __run__, __with__, __apply__, and __also__) that let you work with an object inside a temporary "scope."
* #### __context object__-is the object you’re working on in a scope function
